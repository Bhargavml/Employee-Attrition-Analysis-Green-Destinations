# "Employee Attrition Analysis - Green Destinations"


```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("greendestination (1).csv")
```

```python
# View first rows
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Convert Attrition to binary (Yes=1, No=0)
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
total_employees = len(df)
employees_left = df['Attrition_Flag'].sum()

attrition_rate = (employees_left / total_employees) * 100
attrition_rate
```

```python
age_attrition = df.groupby('Attrition')['Age'].mean()
age_attrition
```

```python
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Age Distribution by Attrition')
plt.show()
```

```python
income_attrition = df.groupby('Attrition')['MonthlyIncome'].mean()
income_attrition
```

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income by Attrition')
plt.show()
```

```python
years_attrition = df.groupby('Attrition')['YearsAtCompany'].mean()
years_attrition
```

```python
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Years at Company by Attrition')
plt.show()
```

```python
sns.countplot(x='Attrition', data=df)
plt.title('Employee Attrition Count')
plt.show()
```

```python
corr = df[['Age', 'MonthlyIncome', 'YearsAtCompany', 'Attrition_Flag']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Key Insights
- Overall attrition rate is approximately **16%**
- Younger employees are more likely to leave
- Lower income is strongly associated with attrition
- Employees with fewer years at the company show higher exit rates

### Business Recommendations
- Improve compensation strategies for early-career roles
- Implement retention programs during first 35 years
- Monitor attrition trends using dashboards
